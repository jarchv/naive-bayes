```python
def mean_and_var(X):
    if X.ndim == 1:
        if X.dtype is not np.float64:
            X = X.astype(np.float64)
        for i in range(N):
            mean += X[i,j]
        mean /= N
            
        for i in range(N):
            diff  = X[i,j] - mean
            var  += diff * diff
        var  /= N
            
        return [mean], [var]
    
    else:
        N     = X.shape[0]
        dim   = X.shape[1]
        means = []
        vars  = []
        
        if X.dtype is not np.float64:
            X = X.astype(np.float64)
        
        for j in range(dim):
            mean = 0.0
            var  = 0.0
            
            for i in range(N):
                mean += X[i,j]
            mean /= N
            
            for i in range(N):
                diff  = X[i,j] - mean
                var  += diff * diff
            var  /= N
            
            means.append(mean)
            vars.append(var)
            
        return means, vars
            
def normalize(X, means, vars):
    N     = X.shape[0]
    
    if X.ndim == 1:
        if X.dtype is not np.float64:
            X = X.astype(np.float64)
            
        for i in range(N):
            X[i] = (X[i] - means) / vars ** 0.5
        
        return X
    
    else:
        N     = X.shape[0]
        dim   = X.shape[1]

        if X.dtype is not np.float64:
            X = X.astype(np.float64)
        
        for j in range(dim):
            for i in range(N):
                X[i, j] = (X[i,j] - means[j]) / vars[j] ** 0.5
        
        return X

        
```

```python
def getClassAttributes(X,y,C):
    
    N    = X.shape[0]
    dist = {c:[] for c in range(C)}
    
    for i in range(N):
        dist[y[i]].append(X[i])
    
    for i in range(C):
        dist[i] = np.array(dist[i], dtype = np.float32)
        
    res  = {}
    for i in range(C):
        means_, vars_ = mean_and_var(dist[i]) 
        res[i] = [(mean, var) for mean, var in zip(means_, vars_)]
    
    return res, dist
```

Function for the Gaussian Probability Distribution

```python
def prob(X, mean, std):
    exp_comp = -(X - mean)**2 /  (2 * std ** 2)
    cnt_comp = 1 / (std * np.sqrt(2 * np.pi))
    return cnt_comp * np.exp(exp_comp)
```

```python
def classProb(input_var, class_mean_var):
    res = {}
    for classVal, class_Att in class_mean_var.items():
        prob_res = 1
        for j in range(len(class_Att)):
            mean_j, var_j =  class_Att[j]
            prob_j_c  = prob(input_var[j], mean_j, var_j**0.5)
            prob_res *= prob_j_c
        res[classVal] = prob_res
    return res
```
